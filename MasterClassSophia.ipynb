## 2.2.Implementation

This function creates the model input through a sliding window algorithm. It takes as input:
- the telemetry
- the number of input of the ML model (sliding window size)

```python
def supervise_input_transform(df, n_input):
    df2 = df.copy()
    for i in range(n_input-1, -1, -1):
        df2['t-'+str(i)] = df['EngineeringValue'].shift(i)
    df2.drop(columns=['EngineeringValue'], inplace=True)
    df2.dropna(axis=0, inplace=True)
    return df2
```

This function creates the model desired output. It takes as input:
- the telemetry
- the input size of the ML model
- the step ahead to predict the telemetry

```python
def supervise_output_transform(df, n_input, N):
    df2 = df.copy()
    df2['t+'+str(N)] = df2['EngineeringValue'].shift(-N)
    df2=df2[n_input-1:]
    df2.dropna(axis=0, inplace=True)
    df2.drop(columns=['EngineeringValue'], inplace=True)
    return df2
```

## 3.2.Model building and training

The model hyperparameters are :
*  Ws : the sliding window size
*  Fs : the number of timesteps ahead to predict



```python
Ws = 400  # Sliding Window Size
Fs = 100   # Steps Ahead to predict 
```

Lets load and prepare the data as seen before:

```python
import tensorflow as tf

# Data is loaded
df = pd.read_csv('https://raw.githubusercontent.com/ThomasGoud/MasterClassSophia/main/data1.csv',index_col="time", parse_dates=['time'])
# First 10000 points are extracted for the training
df = df.iloc[0:10000,:]

# Data is normalized with a StandardScaler()
scaler2 = StandardScaler()
X_std = scaler2.fit_transform(df).squeeze()
df_std = pd.DataFrame(data=X_std, index=df.index, columns=df.columns)

# Data is reshaped for model training
X_train_pd = supervise_input_transform(df_std, Ws)
y_train_pd = supervise_output_transform(df_std, Ws, Fs)

# Size of arrays is adjusted
if len(X_train_pd)>len(y_train_pd):
    X_train_pd = X_train_pd.iloc[:len(y_train_pd),:]
elif len(y_train_pd)>len(X_train_pd):
    y_train_pd=y_train_pd[:len(X_train_pd),:]

# Data is converted to numpy array 
X_train=X_train_pd.loc[:, 't-'+str(Ws-1):'t-0'].to_numpy()
y_train=y_train_pd.loc[:, :'t'+str(Fs)].to_numpy()

# Data is denormalized
y_train_denorm = scaler2.inverse_transform(y_train)
y_train_pd_denorm = pd.DataFrame(data=y_train_denorm, index=y_train_pd.index, columns=y_train_pd.columns)
```

## 4.3.Threshold Computation

In this section we will estimate the signal standard deviation and add/substract n_sig to the predicted value in order to compute the threshold :
- $th_{up} = y_{pred} + n_{\sigma} * \sigma$
- $th_{down} = y_{pred} - n_{\sigma} * \sigma$
- with $\sigma = std(y_{pred})$
